```python
# Quantum Teleportation Algorithm 
# the quantum infomation from one to another
# In classical computer, copy something from one place to another is not a problem.
# But in quantum computer, copying is not allowed. Because the moment you copy actually means you do a measurement
# destroy the quantum state
from qiskit import *
```

```python
circuit = QuantumCircuit(3, 3)
```

```python
%matplotlib inline 
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
```

```python
circuit.cx(1,2)
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
# now the state of q_0 before first barrier, has been teleported to the q_2
# we can verify by measurement
circuit.measure(2, 2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
# the result is c2 c1 c0
# we can notice that the c2 only have the result 1
print(counts)
```
